```python
# Import necessary modules
import numpy as np
import matplotlib.pyplot as plt
from mydiff import forward, backward, central

# Load the data
data = np.loadtxt('testfiles//falling_ball.txt')
time = data[:, 0]
position = data[:, 1]

# Question 1: Plot position vs time
plt.figure(figsize=(10, 6))
plt.plot(time, position, 'b.-')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs Time for Falling Ball')
plt.grid(True)
plt.show()

# Question 2: Calculate velocity using forward and central differences
# Note: mydiff functions return (derivative, x_coordinates)
vel_forward, time_forward = forward(position, time)
vel_central, time_central = central(position, time)

# Plot velocity vs time
plt.figure(figsize=(10, 6))
plt.plot(time_forward, vel_forward, 'r.-', label='Forward difference')
plt.plot(time_central, vel_central, 'b.-', label='Central difference')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs Time')
plt.legend()
plt.grid(True)
plt.show()

# Question 3: Calculate acceleration by differentiating velocity
# Using forward difference on velocity
accel_forward, time_accel_fwd = forward(vel_forward, time_forward)

# Using forward difference on central velocity
accel_central, time_accel_ctr = forward(vel_central, time_central)

# Plot acceleration vs time
plt.figure(figsize=(10, 6))
plt.plot(time_accel_fwd, accel_forward, 'r.-', label='Acceleration (forward on velocity)')
plt.plot(time_accel_ctr, accel_central, 'b.-', label='Acceleration (central velocity)')
plt.axhline(y=-9.8, color='k', linestyle='--', linewidth=2, label='g = -9.8 m/s')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s)')
plt.title('Acceleration vs Time')
plt.legend()
plt.grid(True)
plt.show()

# Calculate mean acceleration
mean_accel_forward = np.mean(accel_forward)
mean_accel_central = np.mean(accel_central)

print(f"Mean acceleration (forward): {mean_accel_forward:.2f} m/s")
print(f"Mean acceleration (central): {mean_accel_central:.2f} m/s")
print(f"\nDifference from g = -9.8 m/s:")
print(f"  Forward: {abs(mean_accel_forward - (-9.8)):.2f} m/s")
print(f"  Central: {abs(mean_accel_central - (-9.8)):.2f} m/s")
```
